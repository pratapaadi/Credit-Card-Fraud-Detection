```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/creditcard.csv')
df.head()
```

```python
df.info()
```

```python
total_Duplicate_Records=df.duplicated().sum()
print("Number of duplicate records:" ,total_Duplicate_Records)
Original_Datapoints=list(df.shape)[0]
print("Total Count of Datapoints with duplicates:" ,Original_Datapoints)
```

```python
df1=df.drop_duplicates()
df1.shape
New_Datapoints=list(df1.shape)[0]
print("Total Count of Datapoints without duplicates:" ,New_Datapoints)
```

```python
numerical_features=df.columns[df1.dtypes!='object']
categorical_features=df.columns[df1.dtypes=='object']
print(numerical_features)
print(categorical_features)
```
